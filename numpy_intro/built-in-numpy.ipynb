```python
import numpy as np
```

```python
x = np.zeros((3, 4), dtype=int)
print(x)
```

```python
print('dtype: {}'.format(x.dtype))
```

```python
x = np.ones((4, 5))
print(x)
```

```python
x = np.full((4, 3), 5)
print(x)
```

```python
x = np.eye(5)
print(x)
```

```python
x = np.diag([10, 20, 30, 40, 50])
print(x)
```

```python
x = np.arange(10)
print(x)
```

```python
x = np.arange(4, 10)
print(x)
```

```python
x = np.arange(1, 14, 3)
print(x)
```

```python
x = np.linspace(0, 25, 10)
print(x)
```

```python
x = np.linspace(0, 25, 10, endpoint=False)
print(x)
```

```python
x = np.arange(20)
print(x)
```

```python
x = np.reshape(x, (4, 5))
print(x)
```

```python
x = np.reshape(x, (2, 10))
print(x)
```

```python
x = np.reshape(x, (10, 2))
print(x)
```

```python
y = np.arange(20).reshape((10, 2))
print(y)
```

```python
x = np.random.random((3, 3))
print(x)
```

```python
y = np.random.randint(4, 15, (3, 2))
print(y)
```

```python
x = np.random.normal(0, 0.1, size=(20, 20))
print(x)
```
