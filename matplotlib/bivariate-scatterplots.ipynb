```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv('database.csv', low_memory=False)
print(fuel_econ.shape)
fuel_econ.head(6)
```

```python
fuel_econ['Combined MPG (FT1)'] = fuel_econ['Combined MPG (FT1)'].astype('O')
fuel_econ.loc[fuel_econ['Combined MPG (FT1)'].notnull(), 'Combined MPG (FT1)'] = fuel_econ.loc[fuel_econ['Combined MPG (FT1)'].notnull(), 'Combined MPG (FT1)'].astype(int)
plt.scatter(data = fuel_econ, x = 'Engine Displacement', y = 'Combined MPG (FT1)')
plt.xlabel('Engine Displacement')
plt.ylabel('Combined MPG (FT1)')
```

```python
sb.regplot(data = fuel_econ, x = 'Engine Displacement', y = 'Combined MPG (FT1)')
plt.xlabel('Engine Displacement')
plt.ylabel('Combined MPG (FT1)')
```

```python
plt.scatter(data = fuel_econ, x = 'Engine Displacement', y = 'Combined MPG (FT1)', alpha = 1/5)
plt.xlabel('Engine Displacement')
plt.ylabel('Combined MPG (FT1)')
```

```python
sb.regplot(data = fuel_econ, x = 'Year', y = 'Combined MPG (FT1)')
plt.xlabel('Year')
plt.ylabel('Combined MPG (FT1)')
```

```python
sb.regplot(data = fuel_econ, x = 'Year', y = 'Combined MPG (FT1)', x_jitter = 0.3)
plt.xlabel('Year')
plt.ylabel('Combined MPG (FT1)')
```

```python
sb.regplot(data = fuel_econ, x = 'Year', y = 'Combined MPG (FT1)', x_jitter = 0.3, scatter_kws = {'alpha': 1/5})
plt.xlabel('Year')
plt.ylabel('Combined MPG (FT1)')
```

```python
sb.regplot(data = fuel_econ, x = 'Year', y = 'Combined MPG (FT1)', x_jitter = 0.3, scatter_kws = {'alpha': 1/20})
plt.xlabel('Year')
plt.ylabel('Combined MPG (FT1)')
```
