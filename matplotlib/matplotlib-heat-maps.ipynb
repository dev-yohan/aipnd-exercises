```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv('database.csv', low_memory=False)
print(fuel_econ.shape)
fuel_econ.head(6)
```

```python
plt.hist2d(data = fuel_econ, x = 'Year', y = 'Combined MPG (FT1)', cmin = 0.5, cmap = 'viridis_r')
plt.colorbar()
plt.xlabel('Year')
plt.ylabel('Combined MPG (FT1)')
```
